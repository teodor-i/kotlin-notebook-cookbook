# Introduce the usage of the new functionality

```kotlin
%use serialization
```

```kotlin
@Serializable
data class Product(
    val id: Int,
    val name: String,
    val category: String,
    val price: Double,
    val inStock: Boolean
)
```

```kotlin
val products = listOf(
    Product(1, "Laptop", "Electronics", 1299.99, true),
    Product(2, "Smartphone", "Electronics", 699.99, true),
    Product(3, "Headphones", "Accessories", 149.99, false),
    Product(4, "Keyboard", "Peripherals", 89.99, true),
    Product(5, "Mouse", "Peripherals", 49.99, true),
    Product(6, "Monitor", "Electronics", 349.99, false),
    Product(7, "Charger", "Accessories", 29.99, true)
)
```

# Filter by category

```kotlin
import org.example.library.where

jsonFlow.where {
    "category" eq "Electronics"
}.toJsonString()
```

# Filter using multiple conditions

```kotlin
jsonFlow.where {
    "price" gt 100.00
    "inStock" eq true
}.toJsonString()
```

# Sort by price

```kotlin
import org.example.library.sortedByNumeric

jsonFlow.sortedByNumeric("price", descending = true).toJsonString()
```

# Select only ID and Name

```kotlin
import org.example.library.map

jsonFlow.map { element ->
    if (element is JsonObject) {
        buildJsonObject {
            element["id"]?.let { put("id", it) }
            element["name"]?.let { put("name", it) }
        }
    } else element
}.toJsonString()
```

# Price statistics

```kotlin
import org.example.library.stats

jsonFlow.stats("price").toJsonString()
```

# Group by category

```kotlin
import org.example.library.groupedBy

jsonFlow.groupedBy("category").toJsonString()
```

# Combining operations

```kotlin
jsonFlow
    .where {
        "category" eq "Electronics"
        "price" gt 500.00
    }
    .stats("price").toJsonString()
```
